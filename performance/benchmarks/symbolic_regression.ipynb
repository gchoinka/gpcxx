```python
import json
import operator
import matplotlib.pyplot as plt
import numpy as np
```

```python
filename="../../build/koza_evolution.json"
```

```python
with open( filename , "r" ) as f:
    evolution = json.load( f )
```

```python
fitness_max = []
fitness_min = []
fitness_mean = []
for generation in evolution:
    generation_max = 0.0
    generation_min = 1.0
    generation_mean = 0.0
    for individuum in generation:
        f = individuum[ "fitness" ]
        generation_max = max( generation_max , f )
        generation_min = min( generation_min , f );
        generation_mean += f
    generation_mean /= float( len( generation ) )
    fitness_max.append( generation_max )
    fitness_min.append( generation_min )
    fitness_mean.append( generation_mean )
```

```python
print fitness_max
print fitness_min
print fitness_mean
```

```python
plt.plot( fitness_max )
plt.plot( fitness_min )
plt.plot( fitness_mean )
plt.axis( [0,50,-0.1,1.1] )
plt.show()
```
